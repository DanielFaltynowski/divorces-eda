# Handle Missing Data

```python
len(divorces)
```

```python
divorces = divorces.dropna(
    subset='education_man'
)
```

```python
len(divorces)
```

```python
divorces['education_man'].value_counts(
    dropna=False
)
```

```python
divorces.loc[:, ['num_kids']] = divorces['num_kids'].fillna(
    value=0
)
```

```python
divorces['num_kids'].value_counts(
    dropna=False
)
```

# Add New Informations

```python
divorces['divorce_date_year'] =pd.to_datetime(divorces['divorce_date']).dt.year.astype(int)
divorces['divorce_date_month'] = pd.to_datetime(divorces['divorce_date']).dt.month.astype(int)
divorces['divorce_date_day'] = pd.to_datetime(divorces['divorce_date']).dt.day.astype(int)

divorces['dob_man_year'] = pd.to_datetime(divorces['dob_man']).dt.year.astype(int)
divorces['dob_man_month'] = pd.to_datetime(divorces['dob_man']).dt.month.astype(int)
divorces['dob_man_day'] = pd.to_datetime(divorces['dob_man']).dt.day.astype(int)

divorces['dob_woman_year'] = pd.to_datetime(divorces['dob_woman']).dt.year.astype(int)
divorces['dob_woman_month'] = pd.to_datetime(divorces['dob_woman']).dt.month.astype(int)
divorces['dob_woman_day'] = pd.to_datetime(divorces['dob_woman']).dt.day.astype(int)

divorces['marriage_date_year'] = pd.to_datetime(divorces['marriage_date']).dt.year.astype(int)
divorces['marriage_date_month'] = pd.to_datetime(divorces['marriage_date']).dt.month.astype(int)
divorces['marriage_date_day'] = pd.to_datetime(divorces['marriage_date']).dt.day.astype(int)

divorces
```

```python
divorces['age_difference'] = divorces['dob_woman_year'] - divorces['dob_man_year']

divorces
```

```python
divorces['num_kids'] = divorces['num_kids'].astype(int)

divorces
```

```python
mapping = {
    'Primary': 1,
    'Preparatory': 2,
    'Secondary': 3,
    'Professional': 4,
    'Other': 0
}

divorces['education_man_numeric'] = divorces['education_man'].map(mapping)
divorces['education_woman_numeric'] = divorces['education_woman'].map(mapping)

divorces['education_man_numeric'] = divorces['education_man_numeric'].astype(int)
divorces['education_woman_numeric'] = divorces['education_woman_numeric'].astype(int)

divorces
```

```python
divorces.to_csv(
    path_or_buf='./../data/divorce_processed.csv'
)
```
